```python
"""
ABTalks Season-3 AI Engineering - Day 20
Task: Embedding Similarity Calculator
Tech Stack: NumPy / SciPy

Objective:
Build a reusable NLP utility to compute similarity between embeddings
using cosine similarity, dot product, and Euclidean distance.
"""

import numpy as np
from scipy.spatial.distance import euclidean


class EmbeddingSimilarityCalculator:
    """
    Professional utility for embedding similarity calculations.
    """

    def __init__(self):
        pass

    @staticmethod
    def validate_vectors(vec1, vec2):
        """
        Validates input vectors.
        """
        if not isinstance(vec1, np.ndarray):
            vec1 = np.array(vec1, dtype=float)

        if not isinstance(vec2, np.ndarray):
            vec2 = np.array(vec2, dtype=float)

        if vec1.shape != vec2.shape:
            raise ValueError("Both embedding vectors must have the same dimensions.")

        if vec1.ndim != 1 or vec2.ndim != 1:
            raise ValueError("Only 1D embedding vectors are supported.")

        return vec1, vec2

    @staticmethod
    def cosine_similarity(vec1, vec2):
        """
        Compute cosine similarity.
        """
        vec1, vec2 = EmbeddingSimilarityCalculator.validate_vectors(vec1, vec2)

        norm1 = np.linalg.norm(vec1)
        norm2 = np.linalg.norm(vec2)

        if norm1 == 0 or norm2 == 0:
            raise ValueError("Zero vector detected. Cosine similarity undefined.")

        similarity = np.dot(vec1, vec2) / (norm1 * norm2)
        return similarity

    @staticmethod
    def dot_product_similarity(vec1, vec2):
        """
        Compute dot product similarity.
        """
        vec1, vec2 = EmbeddingSimilarityCalculator.validate_vectors(vec1, vec2)
        return np.dot(vec1, vec2)

    @staticmethod
    def euclidean_distance(vec1, vec2):
        """
        Compute Euclidean distance.
        """
        vec1, vec2 = EmbeddingSimilarityCalculator.validate_vectors(vec1, vec2)
        return euclidean(vec1, vec2)

    @staticmethod
    def similarity_report(vec1, vec2, label1="Vector A", label2="Vector B"):
        """
        Generate complete similarity report.
        """
        cosine = EmbeddingSimilarityCalculator.cosine_similarity(vec1, vec2)
        dot = EmbeddingSimilarityCalculator.dot_product_similarity(vec1, vec2)
        distance = EmbeddingSimilarityCalculator.euclidean_distance(vec1, vec2)

        print("\n" + "=" * 60)
        print("EMBEDDING SIMILARITY REPORT")
        print("=" * 60)
        print(f"Comparing: {label1} vs {label2}")
        print("-" * 60)
        print(f"Cosine Similarity   : {cosine:.4f}")
        print(f"Dot Product         : {dot:.4f}")
        print(f"Euclidean Distance  : {distance:.4f}")

        if cosine > 0.85:
            interpretation = "Highly Similar"
        elif cosine > 0.60:
            interpretation = "Moderately Similar"
        elif cosine > 0.30:
            interpretation = "Low Similarity"
        else:
            interpretation = "Not Similar"

        print(f"Interpretation      : {interpretation}")
        print("=" * 60)


def main():
    """
    Demo execution with sample embedding vectors.
    """

    embedding_1 = np.array([0.12, 0.45, 0.33, 0.91, 0.56])
    embedding_2 = np.array([0.10, 0.41, 0.30, 0.88, 0.52])
    embedding_3 = np.array([0.90, 0.15, 0.05, 0.10, 0.20])

    calculator = EmbeddingSimilarityCalculator()

    calculator.similarity_report(
        embedding_1,
        embedding_2,
        label1="Sentence Embedding 1",
        label2="Sentence Embedding 2"
    )

    calculator.similarity_report(
        embedding_1,
        embedding_3,
        label1="Sentence Embedding 1",
        label2="Sentence Embedding 3"
    )


if __name__ == "__main__":
    main()
```
